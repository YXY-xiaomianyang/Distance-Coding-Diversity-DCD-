```python
import numpy as np
from osgeo import gdal

# Parameter settings
path=r"E:\application_test\plant\test\test_plant.dat"
s_win=3       #s_win is the window size
gray_l=16      #gray_l is the number of gray levels
write_img(r'E:\application_test\plant\test\3_16_py.tif', im_proj, im_geotrans, DCD_img)
# Histogram Frequency Function
def cout_frequ(arr):
    key = np.unique(arr)
    results = []
    for i in key:
        v =  arr[ arr == i ].size
        results.append(v)
    return results

# DCD calculated by horizontal skip
def DCD_S(windw, win_s, gray_l):
    Circle_win=np.append(windw.reshape(1,-1),windw.reshape(1,-1))
    temp=[]
    for i in range(1, gray_l+1):
        indx=np.where(Circle_win==i)
        if not indx:
            continue
        diff_indx=np.diff(indx)
        diff_indx=diff_indx[0:int((len(diff_indx)+1)/2)]
        
        temp_i=cout_frequ(diff_indx)
        
        temp=temp+temp_i
    temp=np.array(temp)
    p=np.divide(temp, win_s)
    result=np.sum(-p*np.log(p))
    return result

# DCD calculated by horizontal serpentine
def DCD_SS(windw, win_s, gray_l):
    out = windw.copy()
    out[1::2] = windw[1::2,::-1]
    windw = out.ravel()
    Circle_win=np.append(windw.reshape(1,-1),windw.reshape(1,-1))
    temp=[]
    for i in range(1, gray_l+1):
        indx=np.where(Circle_win==i)
        if not indx:
            continue
        diff_indx=np.diff(indx)
        diff_indx=diff_indx[0:int((len(diff_indx)+1)/2)]
        
        temp_i=cout_frequ(diff_indx)
        
        temp=temp+temp_i
    temp=np.array(temp)
    p=np.divide(temp, win_s)
    result=np.sum(-p*np.log(p))
    return result

# DCD calculated by vertical skip
def DCD_C(windw, win_s, gray_l):
    windw=windw.transpose()
    Circle_win=np.append(windw.reshape(1,-1),windw.reshape(1,-1))
    temp=[]
    for i in range(1, gray_l+1):
        indx=np.where(Circle_win==i)
        if not indx:
            continue
        diff_indx=np.diff(indx)
        diff_indx=diff_indx[0:int((len(diff_indx)+1)/2)]
        
        temp_i=cout_frequ(diff_indx)
        
        temp=temp+temp_i
    temp=np.array(temp)
    p=np.divide(temp, win_s)
    result=np.sum(-p*np.log(p))
    return result

# DCD calculated by vertical serpentine
def DCD_CS(windw, win_s, gray_l):
    windw=windw.transpose()
    out = windw.copy()
    out[1::2] = windw[1::2,::-1]
    windw = out.ravel()
    Circle_win=np.append(windw.reshape(1,-1),windw.reshape(1,-1))
    temp=[]
    for i in range(1, gray_l+1):
        indx=np.where(Circle_win==i)
        if not indx:
            continue
        diff_indx=np.diff(indx)
        diff_indx=diff_indx[0:int((len(diff_indx)+1)/2)]
        
        temp_i=cout_frequ(diff_indx)
        
        temp=temp+temp_i
    temp=np.array(temp)
    p=np.divide(temp, win_s)
    result=np.sum(-p*np.log(p))
    return result

# DCD calculated by circle arrangement 
def DCD_CIRCLE_3(windw, win_s, gray_l):
    a=windw.reshape(1,-1) 
    Circle_win=[a[0,4], a[0,5], a[0,8], a[0,7], a[0,6], a[0,3], a[0,0], a[0,1],a[0,2]]
    Circle_win=np.append(Circle_win,Circle_win)
    temp=[]
    for i in range(1, gray_l+1):
        indx=np.where(Circle_win==i)
        if not indx:
            continue
        diff_indx=np.diff(indx)
        diff_indx=diff_indx[0:int((len(diff_indx)+1)/2)]
        
        temp_i=cout_frequ(diff_indx)
        
        temp=temp+temp_i
    temp=np.array(temp)
    p=np.divide(temp, win_s)
    result=np.sum(-p*np.log(p))
    return result

def DCD_CIRCLE_5(windw, win_s, gray_l):
    a=windw.reshape(1,-1)
    Circle_win=[a[0,12], a[0,13], a[0,14], a[0,19], a[0,18], a[0,24], a[0,23], a[0,17],a[0,22],a[0,21], a[0,16], a[0,20], a[0,15], a[0,11], a[0,10], a[0,5], a[0,6],a[0,0],a[0,1], a[0,7], a[0,2], a[0,3], a[0,8], a[0,4],a[0,9]]
    Circle_win=np.append(Circle_win,Circle_win)
    temp=[]
    for i in range(1, gray_l+1):
        indx=np.where(Circle_win==i)
        if not indx:
            continue
        diff_indx=np.diff(indx)
        diff_indx=diff_indx[0:int((len(diff_indx)+1)/2)]
        
        temp_i=cout_frequ(diff_indx)
        
        temp=temp+temp_i
    temp=np.array(temp)
    p=np.divide(temp, win_s)
    result=np.sum(-p*np.log(p))
    return result

def DCD_CIRCLE_7(windw, win_s, gray_l):
    a=windw.reshape(1,-1)
    print(a[0])
    Circle_win=[a[0,24], a[0,25], a[0,26], a[0,27], a[0,34], a[0,33], a[0,41], a[0,48],a[0,40],a[0,32], a[0,47], a[0,39], a[0,46], a[0,45], a[0,38], a[0,31], a[0,44],a[0,37],a[0,43], a[0,42], a[0,36], a[0,30], a[0,35], a[0,29],a[0,28], a[0,21], a[0,22], a[0,23], a[0,14], a[0,7], a[0,15], a[0,16],a[0,8],a[0,0], a[0,1], a[0,9], a[0,2], a[0,3], a[0,10], a[0,17], a[0,4],a[0,11],a[0,5], a[0,6], a[0,12], a[0,18], a[0,13], a[0,19], a[0,20]]
    Circle_win=np.append(Circle_win,Circle_win)
    print(Circle_win)
    temp=[]
    for i in range(1, gray_l+1):
        indx=np.where(Circle_win==i)
        if not indx:
            continue
        diff_indx=np.diff(indx)
        diff_indx=diff_indx[0:int((len(diff_indx)+1)/2)]
        
        temp_i=cout_frequ(diff_indx)
        
        temp=temp+temp_i
    temp=np.array(temp)
    p=np.divide(temp, win_s)
    result=np.sum(-p*np.log(p))
    return result

def read_img(filename):
    dataset = gdal.Open(filename)  
    im_width = dataset.RasterXSize  
    im_height = dataset.RasterYSize  
    im_geotrans = dataset.GetGeoTransform()  
    im_proj = dataset.GetProjection()  
    im_data = dataset.ReadAsArray(0, 0, im_width, im_height).astype(np.float64)  
    del dataset  # 关闭对象，文件dataset
    return im_proj, im_geotrans, im_data, im_height, im_width

def write_img(filename, im_proj, im_geotrans, im_data):
    if 'int8' in im_data.dtype.name:
        datatype = gdal.GDT_Byte
    elif 'int16' in im_data.dtype.name:
        datatype = gdal.GDT_UInt16
    else:
        datatype = gdal.GDT_Float32

    if len(im_data.shape) == 3:
        im_bands, im_height, im_width = im_data.shape
    else:
        im_bands, (im_height, im_width) = 1, im_data.shape
        
    driver = gdal.GetDriverByName("GTiff")  
    dataset = driver.Create(filename, im_width, im_height, im_bands, datatype)

    dataset.SetGeoTransform(im_geotrans)  
    dataset.SetProjection(im_proj)  

    if im_bands == 1:
        dataset.GetRasterBand(1).WriteArray(im_data)  
    else:
        for i in range(im_bands):
            dataset.GetRasterBand(i + 1).WriteArray(im_data[i])
    del dataset


length_win=s_win*s_win    
im_proj, im_geotrans, img, im_height, im_width= read_img(path)  
img=np.mean(img,0)
# print(img.shape)
n1= np.size(img,0)
# print(n1)
n2= np.size(img,1)
b=np.max(img)
c=np.min(img)
img = np.floor(img/((b-c)/gray_l))

#DCD calculating
C = np.zeros((n1,n2))
S = np.zeros((n1,n2))
CS = np.zeros((n1,n2))
SS = np.zeros((n1,n2))
CIRCLE = np.zeros((n1,n2))
half = int((s_win-1)/2)

for i in range(half,n1-half):
    # print(i)
    for j in range (half,n2-half):
        C[i,j] = DCD_C(img[i-half:i+half+1,j-half:j+half+1],length_win,gray_l)
        S[i,j] = DCD_S(img[i-half:i+half+1,j-half:j+half+1],length_win,gray_l)
        CS[i,j] = DCD_CS(img[i-half:i+half+1,j-half:j+half+1],length_win,gray_l)
        SS[i,j] = DCD_SS(img[i-half:i+half+1,j-half:j+half+1],length_win,gray_l)
        CIRCLE[i,j] = DCD_CIRCLE_3(img[i-half:i+half+1,j-half:j+half+1],length_win,gray_l)
        DCD_img=(C+S+CS+SS+CIRCLE)/5

        
        
```
